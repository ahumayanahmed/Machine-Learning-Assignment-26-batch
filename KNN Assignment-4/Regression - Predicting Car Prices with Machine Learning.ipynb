```python
df['num-of-doors'].fillna(df['num-of-doors'].mode()[0], inplace=True)

```

```python
df['normalized-losses'].fillna(df['normalized-losses'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python

df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])

```

```python
df.head()
```

```python
x = df.drop('price', axis=1)
y = df['price']
```

```python
x.head()
```

```python
y.head()
```

```python
df.columns = df.columns.str.strip() 
```

```python
df.corr()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

features=['symboling', 'normalized-losses', 'wheel-base', 'length', 'width',
 'height', 'curb-weight', 'engine-size', 'city-mpg', 'highway-mpg']


for feature in features:
    plt.scatter(x[feature], y)
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.title(f'{feature} vs Price')
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
y_train.head()
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
X_train_scaled
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=8 )
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
results = pd.DataFrame({
    'Actual_Price': y_test.values,
    'Predicted_Price': y_pred
})
print(results.head())
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse_knn = mean_squared_error(y_test, y_pred)
r2_knn = r2_score(y_test, y_pred)
```

```python
r2_knn
```

```python
mse_knn
```

```python
knn.score(x_train, y_train)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
y_pred_lr = lr_model.predict(x_test)
```

```python

results = pd.DataFrame({
    'Actual_Price': y_test.values,
    'Predicted_Price': y_pred_lr
})
print(results.head())
```

```python
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)
```
